## Consulta SQL: 
Escriba una consulta SQL para identificar el número de clientes por grupo económico que: 
- Tienen al menos 2 años de antigüedad con Seguros Bolívar. 
- Han tenido siniestros cuyo valor supera el valor de la prima actual de su póliza.

```python
punto_1_df.columns
```

```python
# Si cada fila respresenta un cliente el DISTINCT se quita
# Si 
query = """
SELECT 
    GRUPO_SECTOR,
    COUNT(DISTINCT ID) AS NUM_CLIENTES
FROM 
    punto_1_df
WHERE 
    ANTIGUEDAD_CLIENTE >= 2
    AND NOMBRE_RAMO_EMISION = 'PRODUCTO 24'
    AND LIQUIDADO_DESDE_2019 > PRIMA_EMITIDA
GROUP BY 
    GRUPO_SECTOR
"""
```

```python
result = sqldf(query, locals())
result.head(50)
```
